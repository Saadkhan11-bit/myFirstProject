```python
# splitting data into training and testing sets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train = data.drop(['price', 'date', 'street', 'city',  'statezip', 'country', ], axis=1)
test = data['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size=12, random_state=2)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
pred = regr.predict(x_test)
```

```python
#predictions
pred
```

```python
#checking Score
regr.score(x_test, y_test)
```

```python
data['predicted_price'] = regr.predict(data.drop(['price', 'date', 'street', 'city', 'statezip', 'country'], axis=1))
```
